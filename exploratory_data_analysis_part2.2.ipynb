```python
import pandas as pd

# supress scientific notation in df
pd.options.display.float_format = '{:.0f}'.format
```

```python
# create file paths
path_2000 = 'Data/final_tmdb_data_2000.csv.gz'
path_2001 = 'Data/final_tmdb_data_2001.csv.gz'
```

```python
# load data
df_2000 = pd.read_csv(path_2000)
df_2001 = pd.read_csv(path_2001)
```

```python
df_2000.head()
```

```python
df_2001.head()
```

```python
df_final = pd.concat([df_2000,df_2001])
```

```python
df_final.head()
```

```python
# check duplicates
df_final.duplicated(subset='imdb_id').sum()
```

```python
# drop duplicates
df_final = df_final.drop_duplicates(subset='imdb_id')
df_final.duplicated(subset='imdb_id').sum()
```

```python
# total movies in df
df_final['imdb_id'].nunique()
```

```python
# number of movies with valid financial information
valid_info = df_final[(df_final['budget']>0) | (df_final['revenue']>0)]['imdb_id'].nunique()
print(f'- {valid_info} movies have budget or revenue information')
```

```python
df_valid = df_final[(df_final['budget']>0) | (df_final['revenue']>0)]
df_valid.head()
```

```python
# number of movies in each certification category
df_valid['certification'].value_counts()
```

```python
# average budget and revenue per certification category
df_valid.groupby('certification')[['budget','revenue']].mean()
```
